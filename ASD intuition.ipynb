```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import dio
import asd
import solve
```

```python
nt = 1000
nw = 12
D = dio.temporal_distance(nw)
X = 2*np.random.rand(nt, nw)
```

```python
# rmse = lambda a, b: ((a-b)**2).sum()
# rsq = lambda y, yh: 1 - rmse(y, yh)/rmse(y, y.mean())
getY = lambda B: X.dot(B) + 2*np.random.normal(size=nt)

def estimate(Y, ro, delta, ssq):
    Reg = asd.ASDReg(ro, [(D, delta)])
    mu, _ = asd.MeanCov(X, Y, Reg, ro, ssq)
    return mu

def plot(B, Y, ssq, ros, deltas, ro, delta):
    plt.plot(B, 'o', lw=5, color='0.2')
    plt.plot(xrange(len(B)), np.zeros(len(B)), '--', color='0.9')
    for i, ro in enumerate(ros):
        mu = estimate(Y, ro, delta, ssq)
        plt.plot(mu, '-', lw=2, color=str(i*1.0/len(ros)), label='ro={0}'.format(ro))

    for i, delta in enumerate(deltas):
        mu = estimate(Y, ro, delta, ssq)
        plt.plot(mu, '--', color=[i*0.6/len(deltas) + 0.4, 0.0, 0.0], lw=2, label='ro={0}, d={1}'.format(ro, delta))

    plt.legend(bbox_to_anchor=(1.4, 1.0))
```

```python
B1 = np.ones(nw)
B1[nw/3:2*nw/3] *= -1
Y1 = getY(B1)

ssq = 1.0
delta = 1.0
ros = [0.0, 6.0, 7.0, 10.0]
ro = 2.0
deltas = [0.5, 5.0, 10.0, 30.0]
plot(B1, Y1, ssq, ros, [], ro, delta)
plt.show()
plot(B1, Y1, ssq, [], deltas, ro, delta)
plt.show()
```
